```python
import pulp

model = pulp.LpProblem("Profit_Maximizing_Problem", pulp.LpMaximize)

A = pulp.LpVariable('A', lowBound=0, cat='Continuous')
B = pulp.LpVariable('B', lowBound=0, cat='Continuous')

model += 5000*A + 2500*B

model += 3*A + 2*B <= 20
model += 4*A + 3*B <= 30
model += 4*A + 3*B <= 44

model.solve()

print("Jumlah robot Advance:", A.value())
print("Jumlah robot Basic:", B.value())
print("Profit maksimum:", pulp.value(model.objective))
```

```python
# Latihan 1
# 4x + 3y = 34
# 5x + y = 37
import pulp

model = pulp.LpProblem("Persamaan", pulp.LpMaximize)

x = pulp.LpVariable('x', lowBound=0)
y = pulp.LpVariable('y', lowBound=0)

model += 4*x + 3*y

model += 4*x + 3*y == 34
model += 5*x + y == 37

model.solve()

print("x =", x.value())
print("y =", y.value())
```

```python
# Latihan 2
# 3P + 4B = 11000
# 1P + 7B = 15000
# 2P + 6B = ?

import pulp

model = pulp.LpProblem("Pensil_Buku", pulp.LpMaximize)

P = pulp.LpVariable('P', lowBound=0)
B = pulp.LpVariable('B', lowBound=0)

model += P

model += 3*P + 4*B == 11000
model += 1*P + 7*B == 15000

model.solve()

P_val = P.value()
B_val = B.value()

print("Harga Pensil =", P_val)
print("Harga Buku =", B_val)

print("Harga 2 Pensil dan 6 Buku =", 2*P_val + 6*B_val)
```

```python
# Latihan 3
# Keliling = 44
# L = P-6
# Persamaan 44 = 2(P-6)
import pulp

model = pulp.LpProblem("Persegi_Panjang", pulp.LpMaximize)

P = pulp.LpVariable('P', lowBound=0)
L = pulp.LpVariable('L', lowBound=0)

model += P  # bebas

model += L == P - 6
model += 2*(P + L) == 44

model.solve()

print("Panjang =", P.value())
print("Lebar =", L.value())
```

```python
# Kendala :
# 1x1 + 0x2 <= 16
# 2x1 + 3x2 <= 19
# 1x1 + 1x2 <= 8

from scipy.optimize import linprog

c = [-5, -7]

A = [
    [1, 0],
    [2, 3],
    [1, 1]
]

b = [16, 19, 8]

res = linprog(c, A_ub=A, b_ub=b, bounds=(0, None))

print("x1 =", res.x[0])
print("x2 =", res.x[1])
print("Z Max =", -(res.fun))
```

```python
# Latihan akhir 1
# Carilah nilai x dan y dari persamaan berikut menggunakan algoritma linear programming (SciPy):
# 4x + 12y = 28
# 2x + y = 21

from scipy.optimize import linprog

# minimisasi fungsi 0x + 0y (objective function)
c = [0, 0]

# Equality constraints
# 4x + 12y = 28
# 2x + y = 21
A_eq = [
    [4, 12],
    [2, 1]
]
b_eq = [28, 21]

# x, y >= 0 (non-negativity bounds)
bounds = [(0, None), (0, None)]

sol = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds)

# Check the solution status
if sol.success:
    print("x =", sol.x[0])
    print("y =", sol.x[1])
    # Optional: print the objective function value if c was not all zeros
    # print("Optimal value =", -sol.fun) # For maximization
else:
    print("No feasible solution found.")
    print("Status:", sol.status)
    print("Message:", sol.message)
```

```python
# Latihan akhir 2
# Lisa membuat 3 tas/jam, Muri membuat 4 tas/jam, Total tas = 55, Total jam = 16
# Variabel:
# L = jam kerja Lisa
# M = jam kerja Muri
# Persamaan:
# 3L + 4M = 55
# L + M = 16

from scipy.optimize import linprog

# Fungsi tujuan: minimisasi 0
c = [0, 0]

# kendala persamaan dijadikan <=
A = [
    [3, 4],      # 3L + 4M <= 55
    [-3, -4],    # -3L -4M <= -55

    [1, 1],      # L + M <= 16
    [-1, -1]     # -L - M <= -16
]

b = [55, -55, 16, -16]

bounds = [(0, None), (0, None)]

sol = linprog(c, A_ub=A, b_ub=b, bounds=bounds)

print("Lisa bekerja (jam):", sol.x[0])
print("Muri bekerja (jam):", sol.x[1])
```
